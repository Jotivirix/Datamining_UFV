```python
#importamos las librerias necesarias
import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
```

```python
#cargamos la tabla
file = 'fifa19.csv'
data3 = pd.read_csv (file, na_values = ['0', None, '?'])
#cambiamos los valores vacíos, 0 o interrogaciones por 'missing values'
data3
```

```python
#borramos las columnas con los valores NaN
data3 = data3.dropna(1)
data3
```

## 3.1 ESTANDARIZACIÓN Y NORMALIZACIÓN DE DATOS 

Para evitar que las diferencias de rango en los datos supongan un problema a la hora de procesar la información, el primer paso es estandarizar y  normalizar los datos. Usamos el método StandardScaler de la librería scikit-learn. 
Dado que solo se puede realizar con datos numéricos, seleccionamos las columnas con dichos datos y obviamos las demás (de momento)

```python
#cogemos solo los valores numéricos
dataNum = data3.select_dtypes(include = ["number"])
dataNum
```

```python
from sklearn.preprocessing import StandardScaler #importo StandardScaler de la librería scikit-learn
#transformamos y normalizamos los datos
ss = StandardScaler()
trans = ss.fit_transform(dataNum)
trans
```

## 3.2 AUTOVALORES Y AUTOVECTORES

Los autovalores y autovectores nos permiten caracterizar el espacio de
llegada de la proyección donde vamos a reducir la dimensión de nuestros datos. Estudiamos cuántas dimensiones necesitamos para representar los datos iniciales. Primero obtenemos la matriz de covarianza utilizando el método cov de la librería numpy

```python
# Calculamos la matriz de covarianza
dataC = np.cov(trans.T)
dataC
```

```python
dataC.shape #es una matriz cuadrada
```

Ahora obtendremos los autovalores y autovectores de la matriz utilizando el método linalg.eig también de la librería numpy

```python
# obtenemos los autovalores
eig_vals, eig_vecs = np.linalg.eig(dataC) 

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

Obtenemos el porcentaje de varianza y el acumulado, por cada componente, en un DataFrame y explicaremos lo que representan y las relaciones entre ellos

```python
#DataFrame con el porcentaje de varianza por cada componente
# para el porcentaje de varianza, realizamos una regla de tres, calculando previamente el sumatorio de los valores del vector
print('Porcentaje de varianza \n')
suma = sum(eig_vals) 
num2 = eig_vals*100/suma
num2
    
```

```python
#DataFrame con el porcentaje acumulado por cada componente
print('Porcentaje acumulado por cada componente: \n')
num3 = num2.cumsum()
num3
```

```python
# tabla
tablaFinal = pd.DataFrame([num2, num3])
tablaFinal.T
```

Ahora vamos a explicar qué quieren decir estos datos y cómo se relacion entre ellos: 
Por un lado contamos con los autovalores que nos dicen la cantidad de información que representa cada uno de los componentes. ## 3.4 USANDO LIBRERÍA SCIKIT-LEARN

### 3.4.1 Autovalores y autovectores

Los autovalores y autovectores nos permiten caracterizar el espacio de llegada de la proyección donde vamos a reducir la dimensión de nuestros datos. Estudiamos cuántas dimensiones necesitamos para representar los datos iniciales. Comenzamos spor estandarizar los datos.

```python
colum = ['ID', 'Age', 'Overall', 'Potential', 'Special']
df = dataNum.loc[:, colum].values 
tran = ss.fit_transform(df)
pd.DataFrame(data = tran, columns = colum).head()
```

```python
#instanciamos y aplicamos objeto PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 2) #2D
cPrinN = pca.fit_transform(tran) #componentes principales con nuevas dimensiones
tablaPrin = pd.DataFrame(data = cPrinN, columns = ['Componente 1', 'Componente 2'])
tablaPrin.head()
```

```python
data3[['Nationality']].head()
```

```python
tFin = pd.concat([tablaPrin, data3[['Nationality']]], axis = 1)
tFin.head()
```

```python
print("Forma: ", cPrinN.shape)
porcentaje = pca.explained_variance_ratio_
print(porcentaje)
sum(porcentaje[0:2])
```

Con el resultado obtenido, vemos que con 2 componentes tenemos más del 81% de varianza explicada. Ahora vamos a dibujar la gráfica de dispersión

### Representación gráfica de los elementos del dataset

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)

# plot data
plt.scatter(tFin['Componente 1'], tFin['Componente 2'], alpha=0.5)
plt.axis('equal');

ax.grid()
```
