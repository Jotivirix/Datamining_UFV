### 1) Haz todo el preprocesamiento para crear un set de entrenamiento y otro de validación que permitan crear un clasificador atendiendo sólo a las características necesarias. Explica qué has hecho y porqué. (0,5 puntos)

Necesitamos separar unas instancias de entrenamiento y otras instancias para validación. La división típica suele ser un 80-20 si bien podemos intentar apurar hasta un 75-25 o incluso 70-30 siendo el número más grande el que representa el procentaje del conjunto de entrenamiento. El problema de definir muchos datos para entrenar y pocos para validar se denomina como "overfitting" y nos puede producir un sobreajuste de las instancias del set de validación. Esto no siempre es interesante ya que no estaríamos obteniendo un modelo realista. El ajuste contrario se denomina "underfitting", y del mismo modo que el "overfitting", no produce un modelo fiable al ser menos preciso que la realidad. Necesitamos además separar la clase objetivo, que en este caso será la "Especie"

Lo primero que hemos tenido que hacer, es aleatorizar el dataset ya que los datos vienen ordenados por la clase objetivo, que en nuestro caso es la especie de las flores. De este modo, si no lo reordenásemos de modo aleatorio, al realizar la división 80-20 que hemos pensado, estaríamos entrenando con dos clases objetivo, mientras que para la validación estaríamos buscando solo en una clase, por lo que el set de entrenamiento no quedará bien entrenado pudiendo producir errores de identificación.
